```python
import apyori
```

```python
transactions = [
     ['Milk', 'Bread', 'Saffron'],
     ['Milk', 'Saffron'],
     ['Bread', 'Saffron','Water'],
     ['Bread','Water'],
     ['Milk','Bread'],
     ['Milk','Bread', 'Water'],
     ['Milk','Bread', 'Butter'],
     ['Bread', 'Butter'],
     ['Milk','Bread', 'Butter','Suger'],
        ] 
```

```python
Rules = list(apyori.apriori(transactions, min_support=0.05, min_confidence=0.05))
association_results = list(Rules)
```

```python
print(Rules[0])
```

```python
for i in range(len(Rules)):
    LHS=list(Rules[i][2][0][0])
    RHS=list(Rules[i][2][0][1])
    support=Rules[i][1]
    confidence=Rules[i][2][0][2]
    lift=Rules[i][2][0][3]
    print("LHS:",LHS,"--","RHS:",RHS)
    print("Support:",support)
    print("Confidence:",confidence)
    print("Lift:",lift)
    print(10*"----")
```
